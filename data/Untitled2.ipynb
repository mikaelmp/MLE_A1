```python
def fix_num_of_loans(type_of_loan):
    if pd.isna(type_of_loan) or str(type_of_loan).strip() == "":
        return 0
    return str(type_of_loan).count(",") + 1

# Apply the fix
df['Num_of_Loan'] = df['Type_of_Loan'].apply(fix_num_of_loans).astype(int)
```

```python
df.head()
```

```python
def parse_credit_history_age(s):
    try:
        parts = s.replace(" Years", "").replace(" Months", "").split(" and ")
        years = int(parts[0])
        months = int(parts[1]) if len(parts) > 1 else 0
        return years * 12 + months
    except:
        return pd.NA

df["Credit_History_Total_Months"] = df["Credit_History_Age"].apply(parse_credit_history_age)
```
